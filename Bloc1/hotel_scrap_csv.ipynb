### Sauvegarder booking_results.json en format csv
---

```python
import pandas as pd
import json
import glob

# Chemin vers tes fichiers JSON
json_files = glob.glob("booking_results/results_*.json")  

all_data = []

for file in json_files:
    with open(file, 'r', encoding='utf-8') as f: # ouvr ele fichie rne lecture 'r': read, with : referme le fichier une fois lu
        data = json.load(f) # lit un fichier et le transforme en lsit de dico
        all_data.extend(data)  # ajoute ts les dictionnaires du .json √† la liste all_data

df_hotel = pd.DataFrame(all_data)

df_hotel.to_csv("df_hotel.csv", index=False, encoding="utf-8-sig")
```
