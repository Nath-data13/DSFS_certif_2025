### Import des bibliothéques
---

```python
import os
import boto3
import pandas as pd
from dotenv import load_dotenv
```

### Chargement secrets.env
---

```python
load_dotenv("secrets.env")

# Récupérer les valeurs
AWS_KEY = os.getenv("AWS_ACCESS_KEY_ID")
AWS_SECRET = os.getenv("AWS_SECRET_ACCESS_KEY")
BUCKET_NAME = os.getenv("AWS_BUCKET_NAME")
```

### Initialiser S3 et envoie fichier CSV
---

```python
s3 = boto3.client("s3", 
                  aws_access_key_id=AWS_KEY, 
                  aws_secret_access_key=AWS_SECRET)

# Exemple : envoyer un fichier
s3.upload_file("df_top20_hotel.csv", BUCKET_NAME, "df_top20_hotel.csv")
s3.upload_file("df_meteo.csv", BUCKET_NAME, "df_meteo.csv")
s3.upload_file("df_ville.csv", BUCKET_NAME, "df_ville.csv")
```
