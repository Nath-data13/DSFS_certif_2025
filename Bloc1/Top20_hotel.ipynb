### Import des bibliothéques
---

```python
import pandas as pd
```

### Chargement des données
---

```python
df_top20 = pd.read_csv('df_hotel.csv')
df_ville = pd.read_csv('df_ville.csv')
```

### Création df avec le Top 20 des Hôtels
---

```python
# modifier la colonne rating avec seulement un nombre, new_col = score
df_top20['score']=df_top20['rating'].str.extract(r'(\d+[\.,]?\d*)')[0]

# modifier la virgule en  point pour ensuite faire du SQL et convertir en float
df_top20['score'] = df_top20['score'].str.replace(',', '.').astype(float)

# visu resultat
print(df_top20['score'].head())

# suppression col 'rating' 
df_top20=df_top20.drop(columns=['rating'])

# création df_top20_hotel
liste_top20 = []

for ville in df_top20['ville']:
    trie_ville = df_top20[df_top20['ville'] ==ville] # filtre par ville
    trie_20 = trie_ville.sort_values(by='score').head(20) # trier les 20 premiers meilleurs scores

    liste_top20.append(trie_20)

df_top20_hotel = pd.concat(liste_top20, ignore_index=True)

df_top20_hotel.head()
```

### Ajout id_hotel pour futurs tables SQL + id_ville du df_ville
---

```python
# création id_hotel
df_top20_hotel['id_hotel'] = range(1, len(df_top20_hotel)+1)

# Merge avec df_ville pour récupérer id_ville
df_top20_hotel = df_top20_hotel.merge(
    df_ville[['nom_ville','id_ville']],
    left_on='ville',
    right_on='nom_ville',
    how='left'
)
df_top20_hotel=df_top20_hotel.drop(columns='ville')
df_top20_hotel.head()
```

### Sauvegarde df_top20_hotel en csv
---

```python
df_top20_hotel.to_csv('df_top20_hotel.csv', index=False, encoding='utf-8-sig')
```
