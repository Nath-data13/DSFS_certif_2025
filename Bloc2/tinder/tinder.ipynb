### Quels sont les attributs importants pour un 2√©me rendez-vous ? Les attributs les mieux not√©s pour un 2√©me rendez-vous sont la Sinc√©rit√©, l'Intelligence et l'Humour. Les √©carts les plus importants entre une prise de d√©cision d'un 2√©me rendez-vous pas, sont l‚ÄôAttractivit√©, l‚Äôhumour et les Int√©r√™ts communs. Nous pouvons constater que l‚ÄôAttractivit√©, l‚ÄôHumour et les Interets communs influencent fortement la d√©cision d'un 2eme rendez-vous ou non.


```python
attributes = {
    'attr_o': 'Attractivit√©',
    'sinc_o': 'Sinc√©rit√©',
    'intel_o': 'Intelligence',
    'fun_o': 'Humour',
    'amb_o': 'Ambition',
    'shar_o': 'Int√©r√™ts communs'
}

# fonction pour calculer des moyennes pour chaque d√©cision
def calcule_notes(df, attributes):
    return pd.DataFrame({
        'Crit√®re': list(attributes.values()),
        '2e rendez-vous': [df[df['dec_o'] == 1][col].mean() for col in attributes],
        'Pas de 2e rendez-vous': [df[df['dec_o'] == 0][col].mean() for col in attributes]
    })

notes = calcule_notes(df, attributes)
print(notes)

# restructuration du tableau
notes_melt=notes.melt(id_vars='Crit√®re', var_name='D√©cision', value_name='Note moyenne')
print(notes_melt)

# cr√©ation graph
fig = px.bar(
    notes_melt,
    x='Crit√®re',
    y='Note moyenne',
    color='D√©cision',
    barmode='group',
    title='Notes moyennes par d√©cision de 2e rendez-vous'
)

fig.update_layout(yaxis_range=[0, 9])
fig.show()
```

### Importance des attributs par genre : Hommes vs Femmes / notes moy par d√©cision de 2e rdv. Ce graphique montre que les hommes et les femmes se retrouvent sur l‚Äôimportance de l‚Äôattractivit√© et des int√©r√™ts communs dans le choix d‚Äôun second rendez-vous. En revanche, les femmes semblent plus sensibles √† l‚Äôhumour que les hommes. 

```python

df['gender_HF']=df['gender'].map({0:'Hommes', 1:'Femmes'})

# Filtrer les hommes et les femmes 
df_hommes = df[df['gender_HF'] == 'Hommes']  
df_femmes = df[df['gender_HF'] == 'Femmes']  

# Calcul des moyennes pour les hommes et les femmes
notes_Hommes = calcule_notes(df_hommes, attributes)
notes_Femmes = calcule_notes(df_femmes, attributes)

# Cr√©ation d'une colonne genre
notes_Hommes['gender_HF'] = 'Hommes'
notes_Femmes['gender_HF'] = 'Femmes'

# Restructuration des donn√©es pour les hommes et les femmes
notes_melt_H = notes_Hommes.melt(id_vars=['Crit√®re', 'gender_HF'], var_name='D√©cision', value_name='Note moyenne')
notes_melt_F = notes_Femmes.melt(id_vars=['Crit√®re', 'gender_HF'], var_name='D√©cision', value_name='Note moyenne')

print(notes_melt_H.columns)
print(notes_melt_F.columns)

# Fusion des DataFrames H et F
notes_melt_HF = pd.concat([notes_melt_F, notes_melt_H])

# Affichage du DataFrame
print(f"- DF FUSIONNE 'notes_melt_HF' \n , {notes_melt_HF}")

# cr√©ation du graph

fig = px.bar(
    notes_melt_HF,
    x='Crit√®re',
    y='Note moyenne',
    color='D√©cision',
    barmode='group',
    title='Notes moyennes par d√©cision de 2e rendez-vous (Hommes vs Femmes)',
    facet_row='gender_HF',
    labels={'gender_HF': 'Genre'},
)

fig.update_layout(yaxis_range=[0, 8],
                      width=1000,  # Augmenter la largeur du graphique
                        height=800)  # Augmenter la hauteur du graphique)
fig.show()
```

### Y a t il une relation entre la Nationalit√© et la prise de d√©cisoin d'un 2√©me rendez_vous ? Il semblerait que la proportion la plus √©l√©v√© soit la population Europeen, suivit des Asiatiques.

```python
df['race'] =  df['race'].map({1: 'Afro-Americain',2: 'Europeen', 3:'Latino-Americain',4:'Asiatique', 5:'Amerindien', 6:'Other'})

# cr√©ation graph
fig = px.histogram(
    df,
    x='race',
    color='race',
    title = 'Repartition des origines √©thniques'
    )

fig.update_layout(
    bargap=0.1) 

fig.show()
```

```python
# Mapping des labels pour la d√©cision
label_map = {0: " 2√®me rendez-vous", 1: " Pas de 2√®me rendez-vous"}

# Liste des races uniques
races = df['race'].dropna().unique()

# Boucle pour chaque origine
for race in races:
    df_r = df[df['race'] == race].copy()
    df_r['label_decision'] = df_r['dec_o'].map(label_map)

    fig = px.pie(
        df_r,
        names='label_decision',
        color='label_decision',
        title="Acceptation vs Refus selon l'origine",
        hole=0.4
    )

    # Ajouter le nom de l'origine au centre
    fig.add_annotation(
        text=race,
        showarrow=False,
        font=dict(size=18, color="black"),
        x=0.5,
        y=0.5
    )

    fig.show()
```
